```python

#functions of steps
#for individual chromosome
def fitness_fun(individual_chromosome):
  in_sample=[4,-2,7,5,11,1]
  quality=1/np.abs(44.1-np.sum(individual_chromosome*in_sample))
  return quality
#for the whole six population
def pop_fitness(pop):
  qualities=np.zeros(pop.shape[0])
  for inv_num in range(pop.shape[0]):
    qualities[inv_num]=fitness_fun(pop[inv_num,:])
    return qualities
#selecting parents and using them in the first generation
def select_mating_pool(pop,qualities,num_parents):
  parents=np.empty((num_parents,pop.shape[1]))
  for i in range(num_parents):
    max_qual_idx=np.where(qualities==np.max(qualities))
    max_qual_idx=max_qual_idx[0][0]
    parents[i,:]=pop[max_qual_idx,:]
    qualities[max_qual_idx]=-1
  return parents
# crossover with 3 parents and 3 new offsprings
def crossover(parents,n_inviduals=6):
  new_generation=np.empty((6,6))
  new_generation[0:parents.shape[0],:]=parents
  new_generation[3,0:3]=parents[0,0:3]
  new_generation[3,3:]=parents[1,3:]
  new_generation[4,0:3]=parents[1,0:3]
  new_generation[4,3:]=parents[2,3:]
  new_generation[5,0:3]=parents[2,0:3]
  new_generation[5,3:]=parents[0,3:]
  return new_generation
# mutation to have better fitness
def mutation(population):
  for idx in range(population.shape[0]):
    population[idx,4]=population[idx,4]/2
  return population


```

```python

#implementation
sol_per_pop=6
num_parents_mating=3
new_generation=np.zeros((6,6))
new_generation[0,:]=[2.4,-5,8,-2,5,1.1]
new_generation[1,:]=[-0.4,2.7,5,-1,7,0.1]
new_generation[2,:]=[-1,2,2,-3,2,0.9]
new_generation[3,:]=[4,7,12,6.1,1.4,-4]
new_generation[4,:]=[3.1,4,0,2.4,4.8,0]
new_generation[5,:]=[-2,3,-7,6,3,3]
for i in range(10):
  #fitness for population 
  qualities=pop_fitness(new_generation)
  #generating new generation
  parents=select_mating_pool(new_generation,qualities,num_parents_mating)
  # crossover with 3 parents and 3 new offsprings
  new_generation=crossover(parents,n_inviduals=sol_per_pop)
  # mutation to have better fitness
  new_generation=mutation(new_generation)
#prediction error
in_sample=[4,-2,7,5,11,1]
for k in range(6):
  print(np.sum(new_generation[k,:]*in_sample))
  
```
